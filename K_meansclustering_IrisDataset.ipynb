```python
import numpy as np
import pandas as pd
from sklearn import metrics,datasets,cluster
import matplotlib.pyplot as plt

```

```python
iris=datasets.load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df.head()
```

```python
Centroid_list=[]
No_of_clusters=range(1,11)
for i in No_of_clusters:
  kmeans=cluster.KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
  kmeans.fit(iris_df.values)
  Centroid_list.append(kmeans.inertia_)

plt.plot(No_of_clusters,Centroid_list)
plt.show()
```

```python
kmeans=cluster.KMeans(n_clusters=3,init="k-means++",max_iter=300,n_init=10,random_state=0)
centroid=kmeans.fit_predict(iris_df.values)
centroid
```

```python
plt.scatter(iris_df.values[centroid ==0,0],iris_df.values[centroid ==0,1],s=100,c="red",label="Iris-setosa")
plt.scatter(iris_df.values[centroid ==1,0],iris_df.values[centroid ==1,1],s=100,c="blue",label="Iris-versicolor")
plt.scatter(iris_df.values[centroid ==2,0],iris_df.values[centroid ==2,1],s=100,c="green",label="Iris-virginica")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="yellow",label="Centroids")
plt.legend()
plt.show()
```

```python
for i in range(1,5):
  kmeans=cluster.KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
  centroid=kmeans.fit_predict(iris_df.values)
  plt.title(f"For k={i}")
  plt.scatter(iris_df.values[centroid ==0,0],iris_df.values[centroid ==0,1],s=100,c="red",label="Iris-setosa")
  plt.scatter(iris_df.values[centroid ==1,0],iris_df.values[centroid ==1,1],s=100,c="blue",label="Iris-versicolor")
  plt.scatter(iris_df.values[centroid ==2,0],iris_df.values[centroid ==2,1],s=100,c="green",label="Iris-virginica")
  plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="yellow",label="Centroids")
  plt.legend()
  plt.show()
```
